```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
%matplotlib inline
```

```python
exp8_measure_03_df = pd.read_csv('./csv/exp8_measure_03.csv', sep=',')
```

```python
exp8_measure_03_df
```

```python
exp8_measure_03_cut_df = exp8_measure_03_df.dropna()
```

```python
exp8_measure_03_cut_df = exp8_measure_03_cut_df[exp8_measure_03_cut_df['reference_M_Pa']>0]
```

```python
# ÐŸÐ¾ÑÑ‚Ñ€Ð¾ÐµÐ½Ð¸Ðµ Ð³Ñ€Ð°Ñ„Ð¸ÐºÐ¾Ð²
plt.figure(figsize=(12, 8))
sns.scatterplot(data=exp8_measure_03_cut_df, x='reference_M_Pa', y='HPL1_2_3_raw', label='HPL1')
sns.scatterplot(data=exp8_measure_03_cut_df, x='reference_M_Pa', y='HPL3_2_2_raw', label='HPL3')
sns.scatterplot(data=exp8_measure_03_cut_df, x='reference_M_Pa', y='HPL4_2_0_raw', label='HPL4')
plt.xlabel('Ð”Ð°Ð²Ð»ÐµÐ½Ð¸Ðµ Ð½Ð° ÑÑ‚Ð°Ð»Ð¾Ð½Ð½Ð¾Ð¼ Ð´Ð°Ñ‚Ñ‡Ð¸ÐºÐµ, ÐœÐŸÐ°')
plt.ylabel('Ð—Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ ÐÐ¦ÐŸ, ÐµÐ´.')
plt.title('ÐÐ°Ð¿Ñ€ÑÐ¶ÐµÐ½Ð¸Ðµ vs Ð”Ð°Ð²Ð»ÐµÐ½Ð¸Ðµ Ð½Ð° ÑÑ‚Ð°Ð»Ð¾Ð½Ð½Ð¾Ð¼ Ð´Ð°Ñ‚Ñ‡Ð¸ÐºÐµ')
plt.legend()
plt.savefig(f'./results/fig/exp_data.png')
plt.show()
```

```python
def plot_regression_with_error(sensor_voltage, sensor_name):
    # ÐŸÐ¾Ð´Ð³Ð¾Ñ‚Ð¾Ð²ÐºÐ° Ð´Ð°Ð½Ð½Ñ‹Ñ…
    x = exp8_measure_03_cut_df[sensor_voltage].values.reshape(-1, 1)
    y = exp8_measure_03_cut_df['reference_M_Pa'].values

    # Ð¡Ð¾Ð·Ð´Ð°Ð½Ð¸Ðµ Ð¸ Ð¾Ð±ÑƒÑ‡ÐµÐ½Ð¸Ðµ Ð¼Ð¾Ð´ÐµÐ»Ð¸ Ð»Ð¸Ð½ÐµÐ¹Ð½Ð¾Ð¹ Ñ€ÐµÐ³Ñ€ÐµÑÑÐ¸Ð¸
    model = LinearRegression()
    model.fit(x, y)

    # ÐŸÐ¾Ð»ÑƒÑ‡ÐµÐ½Ð¸Ðµ ÐºÐ¾ÑÑ„Ñ„Ð¸Ñ†Ð¸ÐµÐ½Ñ‚Ð¾Ð² Ñ€ÐµÐ³Ñ€ÐµÑÑÐ¸Ð¸
    coef = model.coef_[0]
    intercept = model.intercept_

    # ÐŸÑ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð½Ñ‹Ðµ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ
    y_pred = model.predict(x)

    # Ð’Ñ‹Ñ‡Ð¸ÑÐ»ÐµÐ½Ð¸Ðµ ÐºÐ¾ÑÑ„Ñ„Ð¸Ñ†Ð¸ÐµÐ½Ñ‚Ð° Ð´ÐµÑ‚ÐµÑ€Ð¼Ð¸Ð½Ð°Ñ†Ð¸Ð¸ R^2 Ð¸ ÑÑ€ÐµÐ´Ð½ÐµÐ¹ ÐºÐ²Ð°Ð´Ñ€Ð°Ñ‚Ð¸Ñ‡Ð½Ð¾Ð¹ Ð¾ÑˆÐ¸Ð±ÐºÐ¸
    r2 = r2_score(y, y_pred)
    mse = mean_squared_error(y, y_pred)

    # Ð’Ñ‹Ñ‡Ð¸ÑÐ»ÐµÐ½Ð¸Ðµ ÑÑ€ÐµÐ´Ð½ÐµÐ¹ Ð¾Ñ‚Ð½Ð¾ÑÐ¸Ñ‚ÐµÐ»ÑŒÐ½Ð¾Ð¹ Ð¾ÑˆÐ¸Ð±ÐºÐ¸
    relative_errors = np.abs((y - y_pred) / y) * 100
    mean_relative_error = np.mean(relative_errors)

    # ÐŸÐ¾ÑÑ‚Ñ€Ð¾ÐµÐ½Ð¸Ðµ Ð³Ñ€Ð°Ñ„Ð¸ÐºÐ°
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, color='blue', label='Ð”Ð°Ð½Ð½Ñ‹Ðµ Ð¸Ð·Ð¼ÐµÑ€ÐµÐ½Ð¸Ð¹')
    plt.plot(x, y_pred, color='red', label=f'Ð›Ð¸Ð½Ð¸Ñ Ñ€ÐµÐ³Ñ€ÐµÑÑÐ¸Ð¸: y = {coef:.4f}x + {intercept:.4f}')
    plt.xlabel('Ð—Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ ÐÐ¦ÐŸ, ÐµÐ´.')
    plt.ylabel('Ð”Ð°Ð²Ð»ÐµÐ½Ð¸Ðµ Ð½Ð° ÑÑ‚Ð°Ð»Ð¾Ð½Ð½Ð¾Ð¼ Ð´Ð°Ñ‚Ñ‡Ð¸ÐºÐµ, ÐœÐŸÐ°')
    plt.title(f'Ð›Ð¸Ð½Ð¸Ñ Ñ€ÐµÐ³Ñ€ÐµÑÑÐ¸Ð¸. {sensor_name}\n$R^2 = {r2:.4f}$, MSE = {mse:.4f}, MRE = {mean_relative_error:.2f}%')
    plt.legend()
    plt.grid(True)
    plt.savefig(f'./results/fig/regression_{sensor_name}.png')
    plt.show()
    print(f'{sensor_name} y = {coef}x + {intercept}')

    # ÐŸÐ¾ÑÑ‚Ñ€Ð¾ÐµÐ½Ð¸Ðµ Ð³Ñ€Ð°Ñ„Ð¸ÐºÐ° Ð¾ÑÑ‚Ð°Ñ‚ÐºÐ¾Ð²
    residuals = y - y_pred
    plt.figure(figsize=(10, 4))
    plt.scatter(x, residuals, color='purple')
    plt.axhline(y=0, color='black', linestyle='--')
    plt.xlabel('Ð—Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ ÐÐ¦ÐŸ, ÐµÐ´.')
    plt.ylabel('ÐžÑÑ‚Ð°Ñ‚ÐºÐ¸, ÐœÐŸÐ°')
    plt.title(f'Ð“Ñ€Ð°Ñ„Ð¸Ðº Ð¾ÑÑ‚Ð°Ñ‚ÐºÐ¾Ð². {sensor_name}')
    plt.grid(True)
    plt.savefig(f'./results/fig/residuals_{sensor_name}.png')
    plt.show()
```

```python
# ÐŸÐ¾ÑÑ‚Ñ€Ð¾ÐµÐ½Ð¸Ðµ Ð³Ñ€Ð°Ñ„Ð¸ÐºÐ¾Ð² Ð´Ð»Ñ ÐºÐ°Ð¶Ð´Ð¾Ð³Ð¾ Ð´Ð°Ñ‚Ñ‡Ð¸ÐºÐ°
plot_regression_with_error('HPL1_2_3_raw', 'HPL1')
plot_regression_with_error('HPL3_2_2_raw', 'HPL3')
plot_regression_with_error('HPL4_2_0_raw', 'HPL4')
```
